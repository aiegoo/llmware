
 This example shows how to parse .zip archives that include a mix of different file types - as an example,<br>
    we will use a set of Microsoft Investor Relations files, published on the Microsoft IR website -<br>
      --  https://www.microsoft.com/en-us/investor<br>
    We will pull a consolidated set of zip archives from llmware sample files repository that consists of<br>
    zip archives downloaded directly from the Microsoft IR set as the investor kit for each of the quarters<br>
    since 2020.<br>
    Parsing ZIP archives is easy - they are automatically opened, and the source files are then routed to the<br>
    appropriate parser.<br>


```python
from llmware.library import Library
from llmware.configs import LLMWareConfig
from llmware.setup import Setup
from llmware.retrieval import Query
```

  feel free to use postgres or mongo, if installed

```python
LLMWareConfig().set_active_db("sqlite")
```

```python
print("update: downloading and caching microsoft investor relations sample files")
```

```python
microsoft_ir = Setup().load_selected_sample_files(sample_folder="microsoft_ir")
```

```python
print("update: completed downloading files @ files path: ", microsoft_ir)
```

```python
my_lib = Library().create_new_library("microsoft_investor_relations_1")
```

  pass the zip archives like any other file in .add_files method

```python
parsing_output = my_lib.add_files(microsoft_ir)
```

```python
print("update: parsing output: ", parsing_output)
```

  check out the images extracted

```python
print("update: images extracted to path: ", my_lib.image_path)
```

  run a quick query

```python
qr = Query(my_lib).text_query("azure", result_count=10)
```
